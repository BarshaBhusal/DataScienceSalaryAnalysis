```python
#●	Write a python program to find out top 15 jobs. Make a bar graph as well.         
```

```python
top_jobs = df['job_title'].value_counts().head(15) #counting the no of employees of top 15 job_title
top_jobs.name = 'No_of_Employees' #naming the value count to No_of_Employees
top_jobs
```

```python

plt.figure(figsize=(10, 6))
colors = ['skyblue', 'blue','lightblue']

plt.bar(x=top_jobs.index, height=top_jobs.values, color=colors)
plt.xlabel('Job Title',fontsize=12, color='darkred', fontweight='bold')
plt.ylabel('Number of Employees',fontsize=12, color='darkred', fontweight='bold')
plt.title('No of employees in each Job', fontsize=12, color='darkred', fontweight='bold')

plt.xticks(rotation=45, ha='right')# Rotate the x-axis labels for better readability

plt.show()

```
